Разработка полнофункциональной программы медицинской диагностики с использованием нейронных сетей для анализа медицинских изображений — сложная задача, требующая обширных данных, специализированных моделей и знаний в предметной области. Однако я могу предоставить вам упрощенный пример того, как нейронные сети можно использовать для классификации изображений в медицинской диагностике. Обратите внимание, что этот пример предназначен для образовательных целей и не заменяет профессиональную медицинскую консультацию или диагностику.

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input, decode_predictions

# Загрузите предварительно обученную модель MobileNetV2
model = MobileNetV2(weights='imagenet')

# Функция для диагностики медицинского изображения
def diagnose_medical_image(image_path):
    img = image.load_img(image_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    processed_img = preprocess_input(img_array)

    # Предсказать вероятности класса
    predictions = model.predict(processed_img)
    decoded_predictions = decode_predictions(predictions, top=3)[0]

    # Отображение предсказанных меток и вероятностей
    for _, label, probability in decoded_predictions:
        print(f"{label}: {probability*100:.2f}%")

    # Показать изображение
    plt.imshow(img)
    plt.axis('off')
    plt.show()

# пример использования
image_path = 'medical_image.jpg'
diagnose_medical_image(image_path)
```

Предварительно обработанное изображение проходит через модель **MobileNetV2** для прогнозирования. Функция **predict()** возвращает вероятности классов для разных категорий. Функция **decode_predictions()** используется для преобразования вероятностей в удобочитаемые метки.

Предсказанные метки и вероятности печатаются, а исходное изображение отображается с помощью **Matplotlib**.

Обратите внимание, что для реальной программы медицинской диагностики необходим специализированный набор данных, точная настройка модели и тщательная проверка.
